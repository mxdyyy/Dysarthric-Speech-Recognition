```python
import csv

# Path to the CSV file with word-to-file name mapping
csv_file_path = r'D:\Edu\miniproj\dataset\dictionary_UASPEECH.csv'

# Initialize an empty list to store the words
word_list = []

# Read the CSV file and extract the words
with open(csv_file_path, 'r') as csvfile:
    reader = csv.reader(csvfile)
    next(reader)  # Skip the header row
    for row in reader:
        word, _ = row
        word_list.append(word)

# Print the list of words
print("Words extracted from the CSV file:")
for word in word_list:
    print(word)
```

## Img renaming

```python
import shutil

# Path to the base directory containing train and test folders
base_directory = r'D:\Edu\miniproj\dataset\images\Dysarthric'

# Create the output directory if it doesn't exist
output_directory = r'D:\Edu\miniproj\renamed_images'
os.makedirs(output_directory, exist_ok=True)

# Iterate through train and test directories
for dataset_folder in ['Train', 'Test']:
    dataset_path = os.path.join(base_directory, dataset_folder)
    for speaker_folder in os.listdir(dataset_path):
        speaker_path = os.path.join(dataset_path, speaker_folder)
        if os.path.isdir(speaker_path):
            for word_folder in os.listdir(speaker_path):
                word_path = os.path.join(speaker_path, word_folder)
                if os.path.isdir(word_path):
                    image_files = [f for f in os.listdir(word_path) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
                    for i, image_file in enumerate(image_files, start=1):
                        # Construct the new image filename
                        new_image_name = f'{speaker_folder}_{word_folder}{i}.jpg'
                        # Save the image to the output directory
                        output_image_path = os.path.join(output_directory, new_image_name)
                        shutil.copy(os.path.join(word_path, image_file), output_image_path)
                        print(f'Saved {output_image_path}')

print('Image renaming completed!')
```
