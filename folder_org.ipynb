## Renaming Test Labels

```python
import os
import csv

# Path to the directory containing the Dysarthric images
base_directory = r"C:\WORK\images\Dysarthric\Test"

# Path to the CSV file with word-to-folder name mapping
csv_file_path = r"C:\WORK\dictionary_UASPEECH.csv"

# Read the CSV file and create a dictionary mapping FILE NAME to WORD
folder_to_word = {}
with open(csv_file_path, 'r') as csvfile:
    reader = csv.reader(csvfile)
    next(reader)  # Skip the header row
    for row in reader:
        word, folder_name = row
        folder_to_word[folder_name] = word

# Recursively traverse the directory structure
for root, dirs, files in os.walk(base_directory):
    for folder_name in dirs:
        if folder_name in folder_to_word:
            # Get the corresponding word from the dictionary
            new_word = folder_to_word[folder_name]
            # Construct the new folder name
            new_folder_name = os.path.join(root, new_word)
            # Rename the folder
            old_path = os.path.join(root, folder_name)
            os.rename(old_path, new_folder_name)
            print(f'Renamed folder {folder_name} to {new_word}')
        else:
            print(f'No mapping found for folder {folder_name}')

print('Folder renaming completed!')
```

## Image renaming

```python
import shutil

# Path to the base directory containing train and test folders
base_directory = r"C:\WORK\images\Dysarthric"

# Create the output directory if it doesn't exist
output_directory = r"C:\WORK\renamed_images"
os.makedirs(output_directory, exist_ok=True)

# Iterate through train and test directories
for dataset_folder in ['Train', 'Test']:
    dataset_path = os.path.join(base_directory, dataset_folder)
    for speaker_folder in os.listdir(dataset_path):
        speaker_path = os.path.join(dataset_path, speaker_folder)
        if os.path.isdir(speaker_path):
            for word_folder in os.listdir(speaker_path):
                word_path = os.path.join(speaker_path, word_folder)
                if os.path.isdir(word_path):
                    image_files = [f for f in os.listdir(word_path) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
                    for i, image_file in enumerate(image_files, start=1):
                        # Construct the new image filename
                        new_image_name = f'{speaker_folder}_{word_folder}{i}.jpg'
                        # Save the image to the output directory
                        output_image_path = os.path.join(output_directory, new_image_name)
                        shutil.copy(os.path.join(word_path, image_file), output_image_path)
                        print(f'Saved {output_image_path}')

print('Image renaming completed!')
```

```python
import os
from PIL import Image
import numpy as np
import tensorflow as tf

def make_spec_ds(data_directory):
    # Initialize lists to store the spectrograms and labels
    spectrograms = []
    labels = []

    # Iterate over all subdirectories in the data directory
    for label_folder in os.listdir(data_directory):
        label_path = os.path.join(data_directory, label_folder)
        if os.path.isdir(label_path):
            # Iterate over all files in the subdirectory
            for file in os.listdir(label_path):
                # Check if the file is an image file
                if file.lower().endswith(('.png', '.jpg', '.jpeg')):
                    # Read the image file
                    img = Image.open(os.path.join(label_path, file))
                    spectrogram = np.array(img)

                    # Append the spectrogram and label to the lists
                    spectrograms.append(spectrogram)
                    labels.append(label_folder)

    # Convert the lists to TensorFlow Datasets
    spectrogram_ds = tf.data.Dataset.from_tensor_slices(spectrograms)
    label_ds = tf.data.Dataset.from_tensor_slices(labels)

    # Combine the spectrogram and label datasets into a single dataset of tuples
    ds = tf.data.Dataset.zip((spectrogram_ds, label_ds))

    return ds
```

```python
def plot_spectrogram(spectrogram, ax):
  if len(spectrogram.shape) > 2:
    assert len(spectrogram.shape) == 3
    spectrogram = np.squeeze(spectrogram, axis=-1)
  # Convert the frequencies to log scale and transpose, so that the time is
  # represented on the x-axis (columns).
  # Add an epsilon to avoid taking a log of zero.
  log_spec = np.log(spectrogram.T + np.finfo(float).eps)
  height = log_spec.shape[0]
  width = log_spec.shape[1]
  X = np.linspace(0, np.size(spectrogram), num=width, dtype=int)
  Y = range(height)
  ax.pcolormesh(X, Y, log_spec)
```
